# num_sold estimation by using LinearRegression

# data analysis

```python
import pandas as  pd
df=pd.read_csv("train.csv")

df1=pd.read_csv("test.csv")
```

```python
df.head()
```

```python
df1.head()
```

```python
df["num_sold"]=df["num_sold"].fillna(df["num_sold"].mean())
```

```python
df=df.drop("id",axis=1)
df1=df1.drop("id",axis=1)
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
import datetime as dt
```

```python
df["year"]=df["date"].dt.year
```

```python
df["month"]=df["date"].dt.month
```

```python
df["day"]=df["date"].dt.day
```

```python
df1["date"]=pd.to_datetime(df1["date"])
```

```python
df1["year"]=df1["date"].dt.year
```

```python
df1["month"]=df1["date"].dt.month
```

```python
df1["day"]=df1["date"].dt.day
```

```python
df=df.drop("date",axis=1)
```

```python
df1=df1.drop("date",axis=1)
```

```python
df=pd.get_dummies(df)
df1=pd.get_dummies(df1)
```

```python
df.corr(numeric_only=True)
```

```python
l=[]
for i in df.corr(numeric_only=True):
    if -0.1<df.corr(numeric_only=True)[i][0]<0.1:
        l.append(i)






df=df.drop(l,axis=1)
```

```python
l=[]
for i in df1.columns:
    if i not in df.columns:
        l.append(i)


df1=df1.drop(l,axis=1)
```

```python
df.shape
```

```python
df=df.sample(4000)
```

```python
l=[]
for i in df.columns:
    if i !="num_sold":
        if i not in df1.columns:
            l.append(i)


df=df.drop(l,axis=1)
```

```python
x=df.drop("num_sold",axis=1)
```

```python
y=df[["num_sold"]]
```

# model training part

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
model1=lr.fit(x_train,y_train)
pred=model1.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(pred,y_test)
```

# we estimated as 0.56
